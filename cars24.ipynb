```python
# Import data from CSV file

import pandas as pd

# Use a raw string for the file path
df = pd.read_csv(r"C:\Users\HP\Downloads\UPGRAD\Batch - 11 - 01\Mentorship\used_cars.csv")

# Display the DataFrame
df
```

```python
#Data Frame informatin

df.info()
```

```python
# Remove unnecessary column

df = df.drop('Details' , axis = 1)
df
```

```python
# Check Null Values

df.isnull().sum()
```

```python
#Drop Null Values

df = df.dropna()
df.shape
```

```python
# Value counts for categorical features

print(df['Transmission'].value_counts())
print(df['Fuel Type'].value_counts())
print(df['Engine Capacity'].value_counts())
print(df['Fuel Type'].value_counts())
```

```python
# Remove rows where 'Transmission' is 'R' or '4'

df = df[~df['Transmission'].isin(['R' ,'4'])]
df.shape
```

```python
# Remove rows where 'Engine Capacity' is 'cc'

df  = df[df['Engine Capacity'] != 'cc']
df.shape
```

```python
# Correct the Data type of column string to int

df['Engine Capacity'] = df['Engine Capacity'].astype(int)
```

```python
# check rows where 'Price(INR)' is 'call'

df[df['Price(INR)'] == 'Call']
```

```python
# Remove rows where 'Price(INR)' is 'call'

df.drop(df[df['Price(INR)']== 'Call'].index , inplace = True)
df.shape
```

```python
# Correct the Data type of column string to int

df['Price(INR)'] = df['Price(INR)'].astype(int)
```

```python
# Check Duplicate rows

df.duplicated().sum()
```

```python
# Remove duplicate row

df = df.drop_duplicates()
df.shape
```

```python
# Count how many rows have Engine Capacity between 600 and 7000
valid_range_count = df[(df['Engine Capacity'] >= 600) & (df['Engine Capacity'] <= 7000)].shape[0]

print("Number of rows with engine capacity between 600 and 7000 cc:", valid_range_count)
```

```python
# Remove rows with unrealistic engine capacities

df = df[(df['Engine Capacity'] >= 600) & (df['Engine Capacity'] <= 7000)]
```
